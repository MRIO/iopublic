```python

import json
import gzip
import base64
import pandas as pd
import matplotlib as mpl
from matplotlib import cm
%matplotlib inline
```

```python
%%time
out = {}
with open('raw') as f:
    for line in f:
        sim, key, val = line.split()
        val = float(val)
        if key == 'decay1ms' and val == 1000.0:
            continue
        if sim not in out:
            out[sim] = json.loads(base64.urlsafe_b64decode(sim))
        out[sim][key] = val

for key, data in out.items():
    selected = data['selected']
    fn_tuned = f'../tuned_networks/{selected}'
    tuning = json.load(open(fn_tuned))
    #fn_network = f'../{tuning["network"]}.gz'
    #network = json.load(gzip.open(fn_network))
    tuning = {k: v for k, v in tuning.items() if not isinstance(v, (dict, list))}
    out[key].update(tuning)
    
df = pd.DataFrame(out.values())
df
```
